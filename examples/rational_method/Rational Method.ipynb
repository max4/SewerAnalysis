# Rational Method

For roadway drainage design, rational method is one of the most commonly used equations for the calculation of peak flow from small areas. [HEC 22](https://www.fhwa.dot.gov/engineering/hydraulics/pubs/10009/10009.pdf) has a good discussion how it work.

A few things to highlight,

- Tc the time of concentration is the time it takes to activate the whole watershed
- Tc is used to get the rainfall intensity from the IDF (intensity-duration-frequency) curve as the duration
- should only be applied to small area < 200 ac
- Only for peak flow estimation



## Rainfall IDF

POINT PRECIPITATION FREQUENCY (PF) ESTIMATES can be found using the online [NOAA Atlas 14](https://hdsc.nws.noaa.gov/hdsc/pfds/) webstie.
