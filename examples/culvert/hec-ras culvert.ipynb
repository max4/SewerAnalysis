# Culvert in HECRAS

HECRAS should give very similar culvert head water results as it implements FHWA standard equations. A review of the [reference manual](https://www.hec.usace.army.mil/software/hec-ras/documentation/HEC-RAS%205.0%20Reference%20Manual.pdf) are shown below



## Culvert Setup

Culvert setup is similar to bridges. It uses 4 sections.

- 3 contraction section 1:1 contraction rate, also this is where the entrance loss is applied.
- 4 expansion section: 1.5:1 expansion rate




## Terminology

- HW: energy grade line, this is different from the terminology in HDS5, which refers to WSU (water surface, HGL)
- TW: energy grade line
- WSU: water depth on entrance/ section 3

## Inlet Control/Outlet Control

Inlet control happens when the the capacity of the culvert pipe is bigger than the flow the inlet can accept. The flow will go through the inet and transition into subcritical flow through the culvert.

In this case, the inlet behaves as a sluice gate or weir and an orifice if flowing full at the inlet.


Outlet control flow occurs when the culvert flow capacity is limited by downstream conditions or by culvert capacity.

HECRAS pointed some limitations of the FHWA nomographs,

- accuracy within 10%
- nomographs assuming a culvert slope of 0.02 ft/ft slope
- HEC-RAS should be more accurate than the nomographs



## Hydraulic Jump

HECRAS can handle hydraulic jump inside the culvert. And if the jump will fill the barrel, it will switch to outlet control assuming the "burp" wouldn't sustain.
