# RTK Method

## Continuous Simulation
To understand the I/I dynamics of a good size collection system, continuous flow monotoring and calibration can be a much better tool than just calibrating the model during the spring.

The seasonal dynamics of the ground water, evaporation and infiltration play significant role in the among of I/I getting into the system.

And that can present big challenges for the modeling team, this [City of Columbus paper](https://www.chijournal.org/R241-12) showed one way to get around it,
- Using seasonal RTK parameters, grwoth vs dormant
- Using monthly IA parameters

A good way to evaluate the seasonal variation of I/I is using the Cumulative Rainfall(in) vs I/I (in). As shown below, there are 3 distinct parts in the chart below,
- from 4/29-12/3, the same amount of rainfall generates less I/I
- from 12/3-4/29, more I/I is observed

So the idea is that for the two different seasons, differernt RTK should be used for calibration. Using monthly IA makes a lot of sense, since it is mostly account for how soil restores its capacility and should have high correlation to the monthly weather patterns. However, the authors caution due to the lack of storms in each month, it might need multiple years of data to fully calibrate such parameters.

## Strategy on Calibrating Long Term Simulation

Hazem showed a [stepped approach](https://www.chijournal.org/R236-01) to calibrate RTK parameters.

Intial abstraction can be "observed" if you have good flow monitoring data. For example after a long dry period, then it rained a little and registered some responses. And that total rainfall can be used as the estimates of the Ia.


Step 1 calibrate the second event without IA impacts on RTK values



-Step 2 R can be directly calculated since there is no IA
-Step 3 Calibrate TK for that event
-Step 4 Calibrate IA parameters


# Challenges of the RTK Method

A great [review](https://www.chijournal.org/R245-08) listed the following,

- AMC(ANTECEDENT MOISTURE CONDITIONS) is hard to estimate
    - intial abstraction 
    - ground water inflow can only be accounted in DWF, not directly in RDII
- Calibration
    - Time consuming, 20h for each meter (Bennett's book)
- Parameters and Physical Characteristics
    - RTK is a black box, it doesn't rely on any physical characteristics
    - No predicting power on system design to reduce I/I


# Physical Meaning of the 3 RTK Sets

This [paper](https://www.chijournal.org/R241-11) gives great explaination on the meaning of RTK values.

This is a great picture showing how the T, K values are affecting the shape of the final hydrograph.



It also has very pratical advices on how to calibrate RTK values.

A good initial calibration event should,
- No need to calibrate IA parameters
    - winter, soil is saturated and evaporation is low
    - 2nd storm of the back to back storm
- the storm event should be dominated by high intensity rainfall time steps, this will generate the shape of hydrograph we can link to calibration parameters
