# Culvert in SWMM5

[SWMM Reference Manual Volume II—Hydraulics ](https://nepis.epa.gov/Exe/ZyPURL.cgi?Dockey=P100S9AS.txt) covers the culvert calculations in SWMM5 in details.

Source code can be found [here](https://www.openswmm.org/SWMM51014/dwflow-c/dwflow_findConduitFlow?HighlightLines=196).

**The main takeaway is that SWMM5 uses inlet control based on FWHA equations, and uses SWMM5 dynamic wave routing for outlet control.**

# Model Setup

The culvert calculation assumes the water moves slowly at either end of the culvert, which is in general true as the channel is much wider than the culvert.

A SWMM5 model is setup to account for it by,
- upstream node modeled as a storage unit with large area
- downstream modeled as a big box pipe
